```python
ppt.LotFrontage           = ppt.LotFrontage.fillna(ppt.LotFrontage.mean())
ppt.MasVnrType             = ppt.MasVnrType.fillna('None')
ppt.MasVnrArea            = ppt.MasVnrArea.fillna(ppt.MasVnrArea.mean())
ppt.BsmtQual     = ppt.BsmtQual.fillna('Gd')
ppt.BsmtCond       = ppt.BsmtCond.fillna('TA')
ppt.BsmtExposure    = ppt.BsmtExposure.fillna('No')
ppt.BsmtFinType1          = ppt.BsmtFinType1.fillna('Unf')
ppt.BsmtFinType2          = ppt.BsmtFinType2.fillna('Unf')
ppt.Electrical     = ppt.Electrical.fillna('SBrkr')
ppt.FireplaceQu     = ppt.FireplaceQu.fillna('Gd')
ppt.GarageType                = ppt.GarageType.fillna('Attchd')
ppt.GarageYrBlt     = ppt.GarageYrBlt.fillna(2005)
ppt.GarageFinish    = ppt.GarageFinish.fillna('Unf')
ppt.GarageQual        = ppt.GarageQual.fillna('TA')
ppt.GarageCond      = ppt.GarageCond.fillna('TA')
# ppt.Pool_Quality          = ppt.Pool_Quality.fillna('Gd')
# ppt.Fence_Quality         = ppt.Fence_Quality.fillna('MnPrv')
# ppt.Miscellaneous_Feature = ppt.Miscellaneous_Feature.fillna('Shed')
```

```python
ppt.isnull().sum() * 100 / ppt.shape[0]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
ppt[ppt.select_dtypes(include='object').columns] = ppt[ppt.select_dtypes(include='object').columns].apply(le.fit_transform)
```

```python
from sklearn.model_selection import train_test_split
ppt_train, ppt_test = train_test_split(ppt, test_size = 0.2)
```

```python
ppt_train_x = ppt_train.iloc[:, 0: -1]
ppt_train_y = ppt_train.iloc[:,    -1]

ppt_test_x = ppt_test.iloc[:, 0: -1]
ppt_test_y = ppt_test.iloc[:,    -1]
```

```python
from sklearn import preprocessing
ppt_train_x = preprocessing.scale(ppt_train_x)
ppt_test_x = preprocessing.scale(ppt_test_x)
```

```python
from sklearn.linear_model import SGDRegressor
reg_ppt = SGDRegressor()
reg_ppt.fit(ppt_train_x, ppt_train_y)
```

```python
Rsquare = reg_ppt.score(ppt_train_x, ppt_train_y)
Rsquare
```

```python
N = ppt_train_x.shape[0]
K = ppt_train_x.shape[1]
AdRsquare = 1 - (1-Rsquare)*(N-1)/(N-K-1)
AdRsquare
```

```python
pred_train = reg_ppt.predict(ppt_train_x)
pred_test  = reg_ppt.predict(ppt_test_x)
```

```python
err_train = ppt_train_y - pred_train
err_test  = ppt_test_y  - pred_test
```

```python
np.mean(err_train)
```

```python
mape_test = np.mean(np.abs(err_test * 100 / ppt_test_y))
mape_test
```
